## Rename Columns 

```python
df = df.rename(columns = {'Inns' : 'innings_batted','NO' : 'not_outs', 'HS' : 'highest_inns_score', 'BF' : 'balls_faced',
                     'SR' : 'batting_strike_rate', '0' : 'ducks_scored', '4s' : '4_boundary', '6s' : '6_boundary'})
```

## Split Span And Make A Career_Length Column

```python
df[['rookie_year','Fina_year']]=df['Span'].str.split('-', expand=True)
```

```python
df.head(5)
```

```python
df.info()
```

```python
## change columns type 
df['rookie_year'] = df['rookie_year'].astype(int)
df['Fina_year'] = df['Fina_year'].astype(int)
```

```python
df.info()
```

```python
df.head(5)
```

```python
## drop span column 
df.drop(columns='Span', inplace=True)
```

```python
df.head()
```

```python
## make a ccolumn for country 
df['country'] = df['Player'].str.split("(", expand=True)[1]
df['country'] = df['country'].str.replace(')','')
```

```python
df.head()
```

```python
df['Player'] = df['Player'].str.split("(", expand=True)[0]
```

```python
df.head(5)
```

## Career_Length

```python
df['Career_Length'] = df['Fina_year'] - df['rookie_year']
```

```python
df.head(2)
```

## Average Career_length

```python
df['Career_Length'].mean()
```

## Average batting_strike_rate for cricketers who played over 10 years 

```python
df[df['Career_Length'] > 10 ]['batting_strike_rate'].mean()
```

## Number of cricketers who played before 1960 

```python
df[df['rookie_year'] < 1960 ]['Player'].count()
```

## max highest_inns_score by country 

```python
df.groupby(['country'], as_index=False)['highest_inns_score'].max().sort_values(by='highest_inns_score', ascending=False)
```

## Average 100 50 ducks_scored by country 

```python
df.groupby('country')[['100', '50', 'ducks_scored']].mean()
```
