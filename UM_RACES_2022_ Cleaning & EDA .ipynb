##  QUESTIONS

```python
df3.columns
```

```python
## 1 - speed difference by gender in 50km, 50mi ? 
```

```python
df3.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
## 2 - what age groups are the best in the 50m races ?
```

```python
df3.groupby(['athlete_age'], as_index=False)['athlete_average_speed'].agg(['mean', 'count']).sort_values(by = 'mean', ascending = False).query('count > 19')
```

```python
## 2 - what age groups are the worest in the 50m races ?
```

```python
df3.groupby(['athlete_age'], as_index=False)['athlete_average_speed'].agg(['mean', 'count']).sort_values(by = 'mean').query('count <= 19')
```

```python
## The effect of seasons on the data 

# spring 3-5
# summer 6-8
# fall 9-11
# winter 12-2
```

```python
df3.head(3)
```

```python
df3['race_day'] = df3['race_day'].str.replace('.', '/')
```

```python
df3['race_day']  = df3['race_day'].str.replace('.', '')
```

```python
df3['month']  = df3['race_day'].str.split('/', expand=True)[1]
```

```python
df3['month'] = df3['month'].astype(int)
```

```python
df3.head(3)
```

```python
df3['race_season'] = df3['month'].apply(lambda x: 'winter' if x > 11 else 'fall' if x > 8  else 'summer' if x > 5 else 'spring')
```

```python
df3.sample(10)
```

```python
df3.groupby(['race_season'], as_index=False)['athlete_average_speed'].agg(['mean', 'count']).sort_values(by = 'mean', ascending = False).query('count > 19')
```

```python
df3.query('race_length == "50mi"').groupby(['race_season'], as_index=False)['athlete_average_speed'].agg(['mean', 'count']).sort_values(by = 'mean', ascending = False).query('count > 19')
```
