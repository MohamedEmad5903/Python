```python
# apply()

sony = ['PS', 'PS2', 'PS3', 'PS4', 'PSP', 'PSV']
nintendo = ['N64', 'GB', 'GBA', 'GC', 'Wii', 'WiiU', 'DS', 'NES', '3DS', 'SNES']
microsoft = ['XB', 'XB360', 'Xone']
pc = ['PC']
atari = ['2600']
panasonic = ['3DO']
sega = ['DG', 'GG', 'SAT', 'SCD']
```

```python
def category(x):
    if x in sony:
        return 'sony'
    elif x in nintendo:
        return 'nintendo'
    elif x in microsoft:
        return 'microsoft'
    elif x in pc:
        return 'pc'
    elif x in atari:
        return 'atari'
    elif x in panasonic:
        return 'panasonic'
    elif x in sega:
        return 'sega'
    else:
        return 'other'    
```

```python
df['Platform'].apply(category).value_counts()
```

```python
df['Platform_Company'] = df['Platform'].apply(category)
```

## Joining data frames with merge()

```python
boys19 = pd.read_csv(r'E:\Python\Pandas\boys2019.csv')
boys20 = pd.read_csv(r'E:\Python\Pandas\boys2020.csv')
```

```python
boys19
```

```python
boys20
```

```python
pd.options.display.max_rows = 500
```

```python
boys19.merge(boys20, how = 'outer', on ="Name", suffixes=['_2019', '_2020'], indicator=True)
```

```python
boys19.merge(boys20, how = 'inner', on ="Name", suffixes=['_2019', '_2020'])
```

```python
boys19.merge(boys20, how = 'right', on ="Name", suffixes=['_2019', '_2020'], indicator=True)
```

```python
boys19.merge(boys20, how = 'left', on ="Name", suffixes=['_2019', '_2020'], indicator=True)
```

```python
boys_19_20 = boys19.merge(boys20, how = 'outer', on ="Name", suffixes=['_2019', '_2020'], indicator=True)
```
