## RFM Analysis

```python
import datetime as dt
```

```python
retail_df['InvoiceDate'] = pd.to_datetime(retail_df['InvoiceDate'])
retail_df['InvoiceNo'] = pd.to_numeric(retail_df['InvoiceNo'])
```

```python
retail_df['InvoiceDate'].max()
```

```python
today_date = dt.datetime(2011, 12, 12)
```

```python
rfm = retail_df.groupby('CustomerID').agg({'InvoiceDate': lambda date: (today_date - date.max()).days,
                                          'InvoiceNo': lambda num: num.nunique(),
                                          'Total_Price': lambda TotalPrice: TotalPrice.sum()})
rfm
```

```python
rfm.columns=['Recency', 'Frequency', 'Monetary']
rfm
```

```python
rfm = rfm[rfm['Monetary'] > 0]
rfm
```

```python
rfm.describe().T
```

```python
rfm['Recency_score'] =  pd.qcut(rfm['Recency'], 5,  labels=[5, 4, 3, 2, 1])

rfm['Frequency_score'] =  pd.qcut(rfm['Frequency'].rank(method='first'), 5,  labels=[1, 2, 3, 4, 5])

rfm['Monetary_score'] =  pd.qcut(rfm['Monetary'], 5,  labels=[1, 2, 3, 4, 5])
```

```python
rfm.sample(15)
```

```python
rfm['RFM_SCORE'] = rfm['Recency_score'].astype(str) + rfm['Frequency_score'].astype(str)
```

```python
rfm.sample(10)
```

```python
## Customer Segmentation By RFM_Score

seg_map = {r'[1-2][1-2]' : 'lost_customer',
           r'[1-2][3-4]' : 'at_risk',
           r'[1-2]5' : 'can _not _loose',
           r'3[1-2]' :'about_to_sleep',
           r'33' : 'need attention',
           r'[3-4][4-5]' : 'loyal customer',
           r'41' : 'promising',
           r'51' : 'new_customer',
           r'[4-5][2-3]' : 'loyality_potential',
           r'5[4-5]' : 'MVP_customer'}
           
rfm['Segment'] = rfm['RFM_SCORE'].replace(seg_map, regex=True)

rfm.sample(10)
```
